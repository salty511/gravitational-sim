```python
import pickle
import logging
import concurrent.futures
from astropy.time import Time

times = Time(tracker._datetime)
earth_true = [Horizons(id=399, location='@0', epochs=time.jd) for time in times]
earth_true_v = [0 for _ in earth_true]

def get_vectors(index, request):
		logging.debug(f"Thread {index}: starting lookup")
		try:
			earth_true_v[index] = request.vectors()
			logging.debug(f"Thread {index}: Finished Successfully")
		except Exception as e:
			logging.debug(f"Thread {index}: Exception {e}")

format = "%(asctime)s: %(message)s"
logging.basicConfig(format=format, level=logging.DEBUG,
                        datefmt="%H:%M:%S")

with concurrent.futures.ThreadPoolExecutor(max_workers=10) as executor:
		for index, request in enumerate(earth_true):
			executor.submit(get_vectors, index, request)

with open("earth_true_v", "wb") as fp:
	pickle.dump(earth_true_v, fp)
```

```python
from numpy.random import default_rng
import numpy as np
from astroquery.jplhorizons import Horizons
from astropy.time import Time
import astropy.units as u

time = (3.15e+7)*0.01

epoch = Time('1977-09-05')  # pass date as string (Y,M,D)
final_epoch = epoch + time * u.s

earth_data = Horizons(id=399, location='@0', epochs=final_epoch.jd, id_type = 'majorbody')
earth_data_v = earth_data.vectors()

AU = 1.496e+11

rng = default_rng()

n_ast = 50

ast_data = rng.standard_normal((n_ast, 3))

for i in range(ast_data.shape[0]):
    ast_data[i] = ast_data[i] * (5.2*AU/np.linalg.norm(ast_data[i]))

ast_data_v = np.array([[1, 1, 1, *ast_data[i]] for i in range(n_ast)])

earth_pos = np.array([*earth_data_v["x"], *earth_data_v["y"], *earth_data_v["z"]])

earth_pointing_v = [earth_pos*AU - ast_data_v[i, 3:] for i in range(n_ast)]

earth_pointing_uv = earth_pointing_v/np.linalg.norm(earth_pointing_v, axis=1).reshape(n_ast, 1)

ast_data_v[:, :3] = ast_data_v[:, :3] * earth_pointing_uv * 4e6
```

```python
%%time

import numpy as np
import copy
import Body as b
import matplotlib.pyplot as plt
from astropy.coordinates import SkyCoord
from astropy.time import Time
from astroquery.jplhorizons import Horizons
import mpl_toolkits.mplot3d.axes3d as p3
from matplotlib import animation
import pickle
from MassTracker import MassTracker



AU = 1.496e+11
G = 6.67e-11
Me = 6e+24
Ms = 1.99e+30
Mm = 6.4169e+23
Mj = 1.9e+27
Msa = 95.16*Me
Mu = 14.54*Me
Mn = 17.15*Me
Re = 6378000
Rs = 109*Re


day_s = 24*60*60
time = (3.15e+7)*5
dt = time/10000

epoch = Time('2000-01-01')  # pass date as string (Y,M,D)

earth_data = Horizons(id=399, location='@0', epochs=epoch.jd, id_type = 'majorbody')
earth_data_v = earth_data.vectors()
sun_data = Horizons(id=10, location='@0', epochs=epoch.jd, id_type = 'majorbody')
sun_data_v = sun_data.vectors()
mars_data = Horizons(id=499, location='@0', epochs=epoch.jd, id_type = 'majorbody')
mars_data_v = mars_data.vectors()
jupiter_data = Horizons(id=599, location='@0', epochs=epoch.jd, id_type = 'majorbody')
jupiter_data_v = jupiter_data.vectors()
saturn_data = Horizons(id=699, location='@0', epochs=epoch.jd, id_type = 'majorbody')
saturn_data_v = saturn_data.vectors()
uranus_data = Horizons(id=799, location='@0', epochs=epoch.jd, id_type = 'majorbody')
uranus_data_v = uranus_data.vectors()
neptune_data = Horizons(id=899, location='@0', epochs=epoch.jd, id_type = 'majorbody')
neptune_data_v = neptune_data.vectors()

n_ast = 10

ast_data = [Horizons(orbit_ids[i], location='@0', epochs=epoch.jd, id_type = 'majorbody') for i in range(n_ast)]
ast_data_v = [ast_data[i].vectors() for i in range(n_ast)]

#ast_mass = (4/3) * np.pi * (diameter_from_mag(14))**3 * (2e-9)

earth = b.Body(vx=earth_data_v['vx']*(AU/day_s),vy=earth_data_v['vy']*(AU/day_s),vz=earth_data_v['vz']*(AU/day_s),t=0,x=earth_data_v['x']*AU,y=earth_data_v['y']*AU,z=earth_data_v['z']*AU,m=Me, r=Re)
sun = b.Body(vx=sun_data_v['vx']*(AU/day_s),vy=sun_data_v['vy']*(AU/day_s),vz=sun_data_v['vz']*(AU/day_s),t=0,x=sun_data_v['x']*AU,y=sun_data_v['y']*AU,z=sun_data_v['z']*AU,m=Ms)
mars = b.Body(vx=mars_data_v['vx']*(AU/day_s),vy=mars_data_v['vy']*(AU/day_s),vz=mars_data_v['vz']*(AU/day_s),t=0,x=mars_data_v['x']*AU,y=mars_data_v['y']*AU,z=mars_data_v['z']*AU,m=Mm)
jupiter = b.Body(vx=jupiter_data_v['vx']*(AU/day_s),vy=jupiter_data_v['vy']*(AU/day_s),vz=jupiter_data_v['vz']*(AU/day_s),t=0,x=jupiter_data_v['x']*AU,y=jupiter_data_v['y']*AU,z=jupiter_data_v['z']*AU,m=Mj)
saturn = b.Body(vx=saturn_data_v['vx']*(AU/day_s),vy=saturn_data_v['vy']*(AU/day_s),vz=saturn_data_v['vz']*(AU/day_s),t=0,x=saturn_data_v['x']*AU,y=saturn_data_v['y']*AU,z=saturn_data_v['z']*AU,m=Msa)
uranus = b.Body(vx=uranus_data_v['vx']*(AU/day_s),vy=uranus_data_v['vy']*(AU/day_s),vz=uranus_data_v['vz']*(AU/day_s),t=0,x=uranus_data_v['x']*AU,y=uranus_data_v['y']*AU,z=uranus_data_v['z']*AU,m=Mu)
neptune = b.Body(vx=neptune_data_v['vx']*(AU/day_s),vy=neptune_data_v['vy']*(AU/day_s),vz=neptune_data_v['vz']*(AU/day_s),t=0,x=neptune_data_v['x']*AU,y=neptune_data_v['y']*AU,z=neptune_data_v['z']*AU,m=Mn)

ast_arr = [b.Body(vx=ast_data_v[i]['vx']*(AU/day_s),vy=ast_data_v[i]['vy']*(AU/day_s),vz=ast_data_v[i]['vz']*(AU/day_s),t=0,x=ast_data_v[i]['x']*AU,y=ast_data_v[i]['y']*AU,z=ast_data_v[i]['z']*AU,m=100000) for i in range(n_ast)]

tracker_nojupiter = MassTracker([earth,sun,mars], time, dt, list_ast=ast_arr, printing = 0, collision=1, method = 2)
tracker_jupiter = MassTracker([earth,sun,mars,jupiter], time, dt, list_ast=ast_arr, printing = 0, collision=1, method = 2)

tracker_nojupiter.propagate(epoch.iso)
tracker_jupiter.propagate(epoch.iso)

with open("tracker_nojupiter.bin", "wb") as f:
    pickle.dump(tracker_nojupiter, f)

with open("tracker_jupiter.bin", "wb") as f:
    pickle.dump(tracker_jupiter, f)


fig = plt.figure(figsize=plt.figaspect(0.5))

ax = fig.add_subplot(1, 2, 1, projection='3d')



ax.plot3D(tracker_nojupiter.track[0,:,4], tracker_nojupiter.track[0,:,5], tracker_nojupiter.track[0,:,6], "b")
ax.plot3D(tracker_nojupiter.track[1,:,4], tracker_nojupiter.track[1,:,5], tracker_nojupiter.track[1,:,6], "g")
ax.plot3D(tracker_nojupiter.track[2,:,4], tracker_nojupiter.track[2,:,5], tracker_nojupiter.track[2,:,6], "r")

for i in range(n_ast):
    ax.plot3D(tracker_nojupiter.track[3+i,:,4], tracker_nojupiter.track[3+i,:,5], tracker_nojupiter.track[3+i,:,6])

ax = fig.add_subplot(1, 2, 2, projection='3d')

ax.plot3D(tracker_jupiter.track[0,:,4], tracker_jupiter.track[0,:,5], tracker_jupiter.track[0,:,6], "b")
ax.plot3D(tracker_jupiter.track[1,:,4], tracker_jupiter.track[1,:,5], tracker_jupiter.track[1,:,6], "g")
ax.plot3D(tracker_jupiter.track[2,:,4], tracker_jupiter.track[2,:,5], tracker_jupiter.track[2,:,6], "r")
ax.plot3D(tracker_jupiter.track[3,:,4], tracker_jupiter.track[3,:,5], tracker_jupiter.track[3,:,6], "r")

for i in range(n_ast):
    ax.plot3D(tracker_jupiter.track[4+i,:,4], tracker_jupiter.track[4+i,:,5], tracker_jupiter.track[4+i,:,6])

plt.show()
```
